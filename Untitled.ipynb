## Regresión lineal

Buscaremos encontrar una predicción de el sueldo (y1), a través de los datos y una regresión lineal

```python
fig, ax = plt.subplots()

ax.scatter(subdf["y2_dias"], subdf["y1"])

plt.show()
```

### Regresión simple

A través de éste gráfico no se puede apreciar relación alguna de estos datos, por lo tanto su regresión no debería tener un coeficiente de determinación cercano a 1.

```python
def lin_reg(X, Y):
    model = LinearRegression()

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=67)
    model.fit(X_train, Y_train)

    r_sq = model.score(X_train, Y_train)
    print("coeficiente de determinación : {} ".format(r_sq))


    print("interceptor: {} ".format(model.intercept_))
    print("inclinacion: {} ".format(model.coef_))

    # Prediciendo
    prediccion = model.predict(X_test)
    print("Respuesta de predicción:\n {} ".format(prediccion))


    # Predicción manual
    pred_manual = model.intercept_ + model.coef_ * X_test
    print("respuesta de la prediccion:\n {}".format(pred_manual))
    return model
```

```python
X = subdf[["y2_hrs"]]
Y = subdf["y1"]
lin_reg(X,Y)
```

Se obtiene un coefiente de determinación de 0.054, lo que da a entender que se tiene cerca del 5% de aciertos al intentar obtener una predicción. Era esperado que este dato aislado no fuera suficiente dado que el scaterplot no denotaba ningúna relación lineal o cercana a una curva.

### Regresión múltiple

Para intentar encontrar un modelo más cercano al generado por los datos, probablemente sea necesario añadir columnas al conjunto de entrenamiento.

```python

X = subdf[["sexo","y2_hrs"]]
Y = subdf["y1"]

X.sexo[X.sexo == "Mujer"] = 1
X.sexo[X.sexo == "Hombre"] = 0
X
```

```python
lin_reg(X,Y)
```

Se puede notar que a mejorado el coeficiente de determinación al aumentar una columna, por lo que, existe una influencia en la característica sexo para determinar y1 (sueldo)

```python
X = subdf[["sexo", "edad" ,"y2_hrs", "y2_dias"]]
Y = subdf["y1"]

X.sexo[X.sexo == "Mujer"] = 1
X.sexo[X.sexo == "Hombre"] = 0
lin_reg(X,Y)
```

Finalmente se agregaron todas las columnas para generar un predictor de más calidad, para predecir el sueldo. Se puede concluir finalmente que no existe una relación lineal para estos vectores seleccionando sueldo en el eje de las ordenadas. Esto no quiere decir que no exista una relación, puede que ésta tenga un carácter más bien polinóminco.

Cómo último intento se ocupo el sexo en el eje Y, solo como prueba, ya que probablemente no funcione debido a la naturaleza de la regresión lineal.

```python
X = subdf[["y1", "edad" ,"y2_hrs", "y2_dias"]]
Y = subdf[["sexo"]]

Y.sexo[Y.sexo == "Mujer"] = 1
Y.sexo[Y.sexo == "Hombre"] = 0
lin_reg(X,Y.sexo)
```

Se puede apreciar que se obtiene un valor menor de determinación, esto sucede ya que, para este probablemente se necesite un predictor de categorías.
